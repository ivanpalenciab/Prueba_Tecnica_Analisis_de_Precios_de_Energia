```python
import requests
import pandas as pd 
from matplotlib import pyplot as plt
from sqlalchemy import create_engine, MetaData,Table, Column
from sqlalchemy.sql.sqltypes import Numeric,Date
```

# Vamos a obtener datos desde la API

```python
url = "https://l2h237eh53.execute-api.us-east-1.amazonaws.com/dev/precios?start_date=2024-03-15&end_date=2024-04-14"
response = requests.get(url)
data = response.json()
```

```python
obteined_data = data.get("data",{})
total_days = data.get("total_days",{})
print(f"el total de dias obtenidos es {total_days}")
```

# Vamos a convertir los datos obtenidos en un dataframe de pandas

```python
row_data = pd.DataFrame.from_dict(obteined_data)
row_data =row_data.reset_index()
row_data = row_data.rename(columns={'index': 'hora'})
row_data = row_data.replace({"24:00":"00:00"})
row_data.head(24)
```

```python
data_long = pd.melt(row_data, id_vars=['hora'], var_name='fecha', value_name='precio')
```

```python
data_long.info()
```

```python
data_long["hora"] = pd.to_datetime(data_long['hora'], format='%H:%M').dt.hour
data_long["fecha"] = pd.to_datetime(data_long["fecha"], format='%Y-%m-%d')
data_long.head()
```

# Buscamos datos nulos y los imputamos

```python
data_long.info()
```

```python
data_long[data_long["precio"].isna()]
```

```python
#Ordenamos por fecha y hora e imputamos con el valor anterior
data_long = data_long.sort_values(by=['fecha', 'hora'])
data_long['precio'] = data_long['precio'].ffill()
```

```python
fechas_unicas = data_long["fecha"].unique()
rango_completo = pd.date_range(start=fechas_unicas.min(), end=fechas_unicas.max(), freq='D')

# Detectamos fecha faltante
fechas_faltantes = rango_completo.difference(fechas_unicas)

print("Fechas faltantes:", fechas_faltantes)
```

Detectamos que no hay datos el dia 18

```python
fecha_inicio = pd.to_datetime('2024-03-15')
fecha_final = pd.to_datetime('2024-03-21')

#Hallaremos el promedio de los 3 dias anteriores y los 3 disas posteriores
data_filtrado = data_long[(data_long['fecha'] >= fecha_inicio) & (data_long['fecha'] <= fecha_final)]
promedios_dias = data_filtrado.groupby(['fecha']).mean()
promedio = promedios_dias.precio.mean()

fecha_faltante = pd.to_datetime("2024-03-18")

lista_horas = data_long["hora"].unique()

nuevos_datos = {}

for i in lista_horas:
    fila = pd.DataFrame({'fecha': [fecha_faltante], 'hora': [i], 'precio': [promedio]})
    data_long = pd.concat([data_long, fila], ignore_index=True)

#Ordenamos por fecha y hora de nuevo
data_long = data_long.sort_values(by=['fecha', 'hora'])
```

# Datos Promedios Diarios

```python
promedios_diarios = data_long.groupby('fecha')['precio'].agg([
        ('promedio', 'mean')
    ]).round(2)
#Promedio movil 7 dias
promedios_diarios['promedio_movil'] = promedios_diarios['promedio'].rolling(window=7).mean()

```

```python
promedios_diarios
```

```python
plt.figure(figsize=(23, 5))
plt.plot(promedios_diarios.index, promedios_diarios["promedio"], label='Promedio')
plt.plot(promedios_diarios.index, promedios_diarios["promedio_movil"], label='Promedio MÃ³vil')

plt.title("Promedios de precios vs Promedio MÃ³vil")
plt.xlabel("Fecha")
plt.ylabel("Promedio")

plt.legend(loc='upper right')
plt.savefig("grafico.png") 

plt.show()
```

Los precios en el periodo tiene un comportamiento creciente.

# Guardamos en la bases de datos

```python
#Configuracion de la base de datos
meta=MetaData()

engine = create_engine("sqlite:///precios.db", echo=True)
conn =engine.connect()

precios = Table("precios", meta,
               Column("fecha",Date ),
               Column("precio_promedio",Numeric(precision=10, scale=2) ),
               Column("precio_7d",Numeric(precision=10, scale=2))
               )

meta.create_all(engine)

for index, row in promedios_diarios.iterrows():
    fecha = index
    promedio = row['promedio']
    promedio_movil = row['promedio_movil']

    query = precios.insert().values(fecha=fecha,precio_promedio=promedio,precio_7d=promedio_movil)
    conn.execute(query)
    conn.commit()
```
